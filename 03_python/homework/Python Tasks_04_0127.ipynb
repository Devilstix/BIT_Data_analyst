2. Sukurkite tuple su klientų vardais. Paverskite jį į sąrašą, pridėkite naują vardą ir vėl
konvertuokite į tuple.

```python
clients = ("Jonas", "Asta", "Tomas")
clients_list = list(clients)
clients_list.append("Milda")
clients = tuple(clients_list)
print(clients)
```

3. Turite sąrašą kainų [10, 20, 15, 30]. Konvertuokite į set ir parodykite, kiek yra unikalių kainų.
Naudokite set() ir len()

```python
prices = [10, 20, 15, 30]
unic_prices = set(prices)
print("Unice prices number:", len(unic_prices))
```

5. Turite list = ["A", "B", "C"] ir set = {"B", "C", "D"}. Raskite bendrus elementus (naudoti set
operaciją: intersection)

```python
list1 = ["A", "B", "C"]
set1 = {"B", "C", "D"}
set_from_list = set(list1)
groupy = set_from_list.intersection(set1)
print("Grouped elements:", groupy)
```

8. Sukurkite list su pirkimo datomis tekstu (pvz. '2023-01-01') ir konvertuokite jas į datetime
objektus naudojant datetime.strptime.

```python
from datetime import datetime
sale_dates = ["2023-01-01", "2023-02-15", "2023-03-10"]
converted_dates = [datetime.strptime(data, "%Y-%m-%d") for data in sale_dates]
print(converted_dates)
```

9. Apskaičiuokite kiek dienų praėjo nuo kiekvienos pirkimo datos iki šiandienos
(datetime.date.today()) ir atspausdinkite.

```python
from datetime import datetime, date
sale_dates = ["2023-01-01", "2023-02-15", "2023-03-10"]
sales_dates_converted = [datetime.strptime(data, "%Y-%m-%d").date() for data in sale_dates]
today = date.today()
for data in sales_dates_converted:
    skirtumas = (today - data).days
    print(f"From {data} is {skirtumas} days apart")
```

10. Sukurkite žodyną, kuriame raktais yra produkto pavadinimai, o reikšmė – sąrašas mėnesinių
pardavimų. Raskite bendrą sumą kiekvienam produktui naudodami for ciklą ir sum().

```python
sales = { "Kompiuteris": [100, 120, 90, 150], "Telefonas": [200, 180, 220, 210], "Planšetė": [80, 95, 100, 85] }
for product, months in sales.items():
    total_sum = sum(months)
    print(f"{product}: {total_sum}")
```

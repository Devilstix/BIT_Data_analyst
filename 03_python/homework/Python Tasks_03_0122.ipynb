7. Patikrinkite, ar dvi datos (2023-03-05 ir 2023-03-20) priklauso tam pačiam mėnesiui

```python
from datetime import date
d1 = date(2023, 3, 5)
d2 = date(2023, 3, 20)
same_day = (d1.year == d2.year) and (d1.month == d2.month)
print(same_day)
```

3. Pabandykite konvertuoti tekstą 'abc' į int ir suvaldykite ValueError.

```python
try:
    number = int("abc")
    print(number)
except ValueError:
    print("Negalima pakeistis raidziu i skaicius")
```

4. Naudokite try/except/finally – net jei yra klaida, finally turi išspausdinti 'Baigta'

```python
try:
    x = 10 / 0
    print("Dalyba galima")
except ZeroDivisionError:
    print("Dalyba negalima")
finally:
    print("Bigta")
```

5. Sukurkite kodą, kuriame vyksta dvi galimos klaidos: dalyba iš 0 ir netinkama konversija.

7. Sukurkite funkciją, kuri grąžina klaidą, jei įvestas skaičius yra neigiamas

```python
def number_lookup(x):
    try:
        if x < 0:
            raise ValueError("Cant be zero")
        return x
    except ValueError as e:
        print("Error: ", e)
print(number_lookup(10))
print(number_lookup(-5))
```

10. Sukurkite ciklą su try/except, kuris bando padalinti 100 iš sąrašo skaičių, tarp kurių yra ir 0.

```python
numbers = [10, 5, 0, 20, 0, 4]
for x in numbers:
    try:
        result = 100 / x
        print(f"100 / {x} = {result}")
    except ZeroDivisionError:
        print("Cant divide by zero")
```

# Užduotys su datetime:

1. Sukurkite šiandienos datą naudojant datetime.date.today() ir ją atspausdinkite

```python
import datetime
time = datetime.date.today()
print(time)
```

2. Sukurkite datą: 2023 m. birželio 1 d. ir atspausdinkite ją

```python
import datetime
time = datetime.date(2023, 6, 1)
print(time)
```

3. Konvertuokite tekstą '2022-12-25' į datetime objektą.

```python
import datetime
text = '2022-12-25'
date = datetime.datetime.strptime(text, '%Y-%m-%d')
print(date)
```

4. Išveskite, kuri savaitės diena buvo 2022-12-25

```python
from datetime import datetime
date = datetime(2022, 12, 25)
weekday = date.strftime("%A")
print(weekday)
```

5. Apskaičiuokite dienų skirtumą tarp 2023-01-01 ir 2023-02-01.

```python
from datetime import date
d1 = date(2023, 1, 1)
d2 = date(2023, 2, 1)
diference = d2 - d1
print(diference)
```

6. Sukurkite ciklą, kuris atspausdina kiekvieno 2023 metų mėnesio 1-ą dieną

```python
from datetime import date
for month in range(1, 13):
    first_day = date(2023, month, 1)
    print(first_day)
```

7. Patikrinkite, ar dvi datos (2023-03-05 ir 2023-03-20) priklauso tam pačiam mėnesiui

```python
from datetime import date
d1 = date(2023, 3, 5)
d2 = date(2023, 3, 20)
same_day = (d1.year == d2.year) and (d1.month == d2.month)
print(same_day)
```

8. Paverskite šiandienos datą į tekstinį formatą: YYYY-MM-DD.

```python
from datetime import date
today = date.today()
formatted = today.strftime("%Y-%m-%d")
print(formatted)
```

9. Sukurkite datą iš metų, mėnesio ir dienos, pateiktų kintamaisiais.

```python
from datetime import date
y = 2026
m = 5
d = 12
newdate = date(y, m, d)
print(newdate)
```

10. Patikrinkite, ar duota data yra savaitgalis (šeštadienis arba sekmadienis)

```python
from datetime import date
d = date(2026, 2, 1)
is_weekend = d.weekday() <= 5
print(is_weekend)
```

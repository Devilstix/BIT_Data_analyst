6. Apskaičiuokite pelną po metų su 5% palūkanomis nuo 1000 €.

```python
pradine_suma = 1000
palukanos = 0.05

galutine_suma = pradine_suma * (1 + palukanos)
pelnas = galutine_suma - pradine_suma

print(pelnas)
```

7. Apskaičiuokite pažymių [7, 8, 9] vidurkį.

```python
pazymiai = [7, 8, 9]
vidurkis = sum(pazymiai) / len(pazymiai)
print(vidurkis)
```

8. Apskaičiuokite, kiek procentų liko negrąžintos paskolos (1000 € – 350 € grąžinta).

```python
paskola = 1000
grazinta = 350
liko_moketi = paskola - grazinta
procent_left = (liko_moketi / paskola) * 100
print(procent_left)
```

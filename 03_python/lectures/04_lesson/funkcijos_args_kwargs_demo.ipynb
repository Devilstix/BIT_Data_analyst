# Python funkcijos: argumentai, *args ir **kwargs

Šiame faile rasite paaiškinimus ir praktinius pavyzdžius apie funkcijų parametrų naudojimą.

## 1. Funkcijos su konkrečiai priskirtais argumentais

```python
def sveikinti(vardas, miestas):
    print(f"Sveiki, aš esu {vardas} iš {miestas}!")

sveikinti("Žydrūnė", "Vilniaus")
```

### Galima naudoti ir argumentų vardus:

```python
sveikinti(miestas="Vilniaus", vardas="Jonas")
```

## 2. Argumentai su numatytosiomis reikšmėmis

```python
def sveikinti(vardas, miestas="Vilnius"):
    print(f"Sveiki, {vardas} iš {miestas}!")

sveikinti("Milda")
sveikinti("Tomas", "Klaipėda")
```

## 3. Funkcijos su `*args`
Naudojama, kai norime perduoti **nežinomą skaičių argumentų** (be vardų).

```python
def suma(*skaiciai):
    print(f"Argumentai: {skaiciai}")
    return sum(skaiciai)

print(suma(1, 2, 3))
print(suma(5, 10, 15, 20))

b = 1 + suma(5, 8, 9, 2.74)
print(b)
```

### `*args` yra tuple – galime per jį eiti su ciklu.

## 4. Funkcijos su `**kwargs`
Naudojama, kai perduodami **vardiniai argumentai** (raktas=reikšmė).

```python
def rodyti_info(**kwargs):
    for raktas, reiksme in kwargs.items():
        print(f"{raktas}: {reiksme}")

rodyti_info(vardas="Jonas", amzius=30, miestas="Panevėžys")
rodyti_info(vardas="Lionia", amzius=70, miestas="Balbieriškis")
```

## 5. Funkcija su visais tipais argumentų

```python
def visa_info(vardas, *hobiai, **kontaktai):
    print(f"Vardas: {vardas}")
    print("Hobiai:", hobiai)
    print("Kontaktai:")
    for raktas, reiksme in kontaktai.items():
        print(f"  {raktas}: {reiksme}")

visa_info("Jonė", "dailė", "joga", "katės", "bėgimas", "kelionės", elpastas="jone45893@gmail.com", miestas="Jonava")

visa_info(1, "du", "Japonija", 3.1415, True, None, elpastas="jone1145893@gmail.com", miestas="Balbieriškis")
```
